```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
sb.set()
```

```python
whr_data = pd.read_csv('WHR2018Chapter2OnlineData.xls - Table2.1.csv')
whr_data.head()
```

```python
whr_data.info()
```

```python
whr_data.describe()
```

```python
whr_data_2014 = whr_data[whr_data['year'] == 2014]
whr_data_2014.head()
```

```python
whr_data_2014_toSingapore = whr_data_2014[whr_data_2014['country'] <= 'Singapore']
whr_data_2014_toSingapore = whr_data_2014_toSingapore.fillna(0)
whr_data_2014_toSingapore.info()
```

```python
print(whr_data_2014_toSingapore)
sb.catplot(data=whr_data_2014_toSingapore, x='Life Ladder' , kind='swarm')
```
